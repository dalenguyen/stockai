```python
import pandas as pd 
import datetime
from loguru import logger
```

```python
from stockai import Stock

td = Stock('MSFT')
prices_list = td.get_historical_prices('2021-01-01', '2021-01-30')

df = pd.DataFrame(prices_list)

# convert time
df['date'] = pd.to_datetime(df['date'], unit='s').dt.strftime('%Y-%m-%d')

df.set_index('date', inplace=True)

df.head()
```

```python
stocks = ['AMZN', 'MSFT', 'INTC', 'FB', 'GOOG']

start = (datetime.datetime.today() - datetime.timedelta(30)).strftime('%Y-%m-%d')
end = datetime.datetime.today().strftime('%Y-%m-%d')

# print(start)
# print(end)

close_prices = pd.DataFrame()
ohlcv_data = {}

# Loop through stock
for ticker in stocks:
    sk = Stock(ticker)
    data = sk.get_historical_prices(start, end)

    ohlcv_data[ticker] = data

#     print(len(data['open']))
    close_prices[ticker] = data['adjclose']
    close_prices['date'] = data['date']

# Convert & set date as index
close_prices['date'] = pd.to_datetime(close_prices['date'], unit='s').dt.strftime('%Y-%m-%d')
close_prices.set_index('date', inplace=True)    
print(close_prices)

print('\nOpen Price for Mircrosoft \n')

print(ohlcv_data['MSFT']['open'])
```

```python
print('\n--Mean\n')
print(close_prices.mean())

print('\n--Medium\n')
print(close_prices.median())

print('\n--std\n')
print(close_prices.std())
```

```python
# Percentage change
daily_return = close_prices.pct_change()
print(daily_return)
```

```python
print('\n--Mean\n')
print(daily_return.mean())
# The best mean daily return is GOOG (0.005698)

print('\n--std\n')
print(daily_return.std())
```

```python
# Rolling mean & standard deviation

## Simple moving avarage (SMA)
print('\n--Mean\n')
print(daily_return.rolling(window=5).mean()) # 5 days

print('\n--Std\n')
print(daily_return.rolling(window=5).std()) # 5 days
```

```python
# Exponential weighted (EW) - Exponential moving average (EMA)
# Since new data carries greater weight, the EMA responds more quickly to price changes than the SMA.

"""
- Exponential moving averages are designed to see price trends over specific time frames like 50 or 200 days.
- Compared to simple moving averages, EMAs give greater weight to recent (more relevant) data.
- Computing the exponential moving average involves applying a multiplier to the SMA.
- Moving average ribbons allow traders to see multiple EMAs at the same time.
"""

print('\n--Mean\n')
print(daily_return.ewm(span=5, min_periods=5).mean()) # 5 days

print('\n--Std\n')
print(daily_return.ewm(span=5, min_periods=5).std()) # 5 days
```
