```python
import pandas as pd 
import datetime
```

```python
from stockai import Stock

td = Stock('MSFT')
prices_list = td.get_historical_prices('2021-01-01', '2021-01-30')

# Delete meta key
del(prices_list['meta'])

# print(prices_list)

df = pd.DataFrame.from_dict(prices_list)
df.head()
```

```python
stocks = ['AMZN', 'MSFT', 'INTC', 'GOOG']

start = (datetime.datetime.today() - datetime.timedelta(30)).strftime('%Y-%m-%d')
end = datetime.datetime.today().strftime('%Y-%m-%d')

# print(start)
# print(end)

cl_price = pd.DataFrame()
ohlcv_data = {}

# Loop through stock
for ticker in stocks:
    sk = Stock(ticker)
    data = sk.get_historical_prices(start, end)
    del(data['meta'])

    ohlcv_data[ticker] = data

#     print(len(data['open']))
    cl_price[ticker] = data['adjclose']
    
print(cl_price)

print('\nOpen Price for Mircrosoft \n')

print(ohlcv_data['MSFT']['open'])
```
